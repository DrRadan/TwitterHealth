```python
import pandas as pd
import numpy as np
import pickle
import plotly.plotly as py 
import pandas as pd 
import copy

  
# some more libraries to plot graph 
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot, plot 
init_notebook_mode(connected = True) 
```

```python
data_2015 = pickle.load(open('../data/df_distance_probs_new.pickle', 'rb'))
data_2015.rename({'probs_act':'probs_activity'}, inplace=True, axis=1)
data_2015 = data_2015.set_index('State')
```

```python
data_2015['timestamp'] = data_2015['timestamp'].apply(lambda x: pd.Timestamp(x, unit='s'))
data_2015['month'] = data_2015['timestamp'].apply(lambda x: x.month)
data_2015_06 = data_2015[(data_2015['month'] == 6) | (data_2015['month'] == 7)][['target','probs_food', 'probs_activity']]
data_2015_11 = data_2015[(data_2015['month'] == 11) | (data_2015['month'] == 12)][['target','probs_food', 'probs_activity']]
```

```python
state_id = pd.read_csv('../data/state.csv', header=None, index_col=[0], names=['state_code','state','num'])
state_id
```

```python
def get_frequency(df):
    num1 = sum(df['target']==1)
    num2 = sum(df['target']==2)
    num3 = sum(df['target']==3)
    return pd.DataFrame([[(num1+num3)/len(df),(num2+num3)/len(df),df['probs_food'].mean(),
                          df['probs_activity'].mean(), len(df)]]
                        ,columns=['food_freq','activity_freq', 'randomforest_food_freq', 
                                  'randomforest_activity_freq', 'num_of_tweet'])
result = data_2015.groupby(lambda x:x)['target','probs_food', 'probs_activity'].apply(get_frequency)
result.reset_index(level=1,drop=True,inplace=True)
result
```

```python
def get_summery_info(data):
    result = data.groupby(lambda x:x)['target','probs_food', 'probs_activity'].apply(get_frequency)
    result.reset_index(level=1,drop=True,inplace=True)
    result = result.merge(state_id, left_index=True, right_index=True)
    return result
```

```python
result = get_summery_info(data_2015_06)
```

```python
def plot_usa_heatmap(result, column_name, colorbar_title, figure_title):
    df = copy.deepcopy(result)
    max_value = max(df[column_name])
    for col in df.columns:
        df[col] = df[col].astype(str)
        
    #scl = ['Greys', 'YlGnBu', 'Greens', 'YlOrRd', 'Bluered', 'RdBu',
    #        'Reds', 'Blues', 'Picnic', 'Rainbow', 'Portland', 'Jet',
     #       'Hot', 'Blackbody', 'Earth', 'Electric', 'Viridis', 'Cividis']
    scl = 'Rainbow'
    #scl = [[0.0, 'rgb(242,240,247)'],[max_value/5, 'rgb(218,218,235)'],[2*max_value/5, 'rgb(188,189,220)'],\
    #            [3*max_value/5, 'rgb(158,154,200)'],[4*max_value/5, 'rgb(117,107,177)'],[5*max_value/5, 'rgb(84,39,143)']]

    data = [ dict(
            type='choropleth',
            colorscale = scl,
            autocolorscale = False,
            locations = df['state_code'],
            z = df[column_name].astype(float),
            locationmode = 'USA-states',
            text = df[column_name],
            marker = dict(
                line = dict (
                    color = 'rgb(255,255,255)',
                    width = 2
                ) ),
            colorbar = dict(
                title = colorbar_title)
            ) ]
    layout = dict(
            title = figure_title,
            geo = dict(
                scope='usa',
                projection=dict( type='albers usa' ),
                showlakes = True,
                lakecolor = 'rgb(255, 255, 255)'),
                 )

    fig = dict( data=data, layout=layout )
    iplot(fig)
    


    
```

# 2015-06 food

```python
plot_usa_heatmap(result, column_name='food_freq', colorbar_title='freq', figure_title='2015-06 food freq heatmap')
plot_usa_heatmap(result, column_name='randomforest_food_freq', colorbar_title='prob', figure_title='2015-06 food prob heatmap')
```
